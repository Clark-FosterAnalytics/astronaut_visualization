```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
pd.set_option('display.expand_frame_repr', False)
%matplotlib inline
```

```python
#Create lists
ages = []
years = []
byears = []
```

```python
#Import data
astronauts = pd.read_csv('NASA astronauts - US ASTRONAUTS.csv')
```

```python
#Add selection years to list
for detail in astronauts["Selection Year"]:
    years.append(detail)
```

```python
#Convert birth dates to years and add to them to list. 
for detail in astronauts["Date of birth"]:
    #Split year from month and date.
    date = detail.split('/')
    #For some astronaut, the entire birth date is recorded. For others, only the birth year is. 
    #The code below tests whether the list we created with the last split includes a month and date or just a year
    #and parses it accordingly. 
    if len(date) > 2:
        byears.append(date[2])
       
    else:
        byears.append(date[0])
```

```python
#Let's take a look at the birth and selection years of the astronauts. 
for i in range(len(years)):
    print("Birth:",byears[i], "Selection:",years[i])
```

```python
#Now, let's determine the ages of the astronauts when they were selected for the space program.

for i in range(len(years)):
    ages.append(years[i] - int(byears[i]))
```

```python
#Stylistic tweaking
sns.set_context("talk")
sns.set_style("dark")
plt.figure(figsize=(16, 6))
#Put selection years and ages at selection in a new data frame and plot it.
age_data = pd.DataFrame({'Selection Year':years ,'Age': ages})
sns.boxplot(x='Selection Year',y='Age',data=age_data).set_title('Ages of Astronauts')
plt.show()
```
